```python
# Step 1
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('hw4_exams.csv')
df.head()
```

```python
# Step 2
df.info()
```

```python
# Step 3
df[['math score', 'reading score', 'writing score']].describe()
```

```python
# Step 4
race_mean_scores = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()
race_mean_scores
```

```python
# Step 5
single_column_df = df[['math score']]
single_column_df.head()
```

```python
# Step 6
single_column_series = df['math score']
single_column_series.head()
```

```python
# Step 8
female_high_math = df[(df['gender'] == 'female') & (df['math score'] >= 90)]
female_high_math
```

```python
df.boxplot(column=['math score', 'reading score', 'writing score'])
```

```python
# Step 9
df.boxplot(column=['math score', 'reading score', 'writing score'])
```

```python
# Step 10
gender_avg_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
gender_avg_scores.plot(kind='bar', rot=0, ylabel='Average Score', xlabel='Gender', title='Average Scores by Gender')
```

```python
# Step 11
sns.set(style='whitegrid')
cat_plot = sns.catplot(x='gender', data=df, kind='count', palette='pastel', height=5, aspect=1.5)
```

```python
# Step 12
cat_plot.set_xticklabels(rotation=45)
```

```python
# Step 13
scatter_specific = sns.scatterplot(x='reading score', y='writing score', hue='test preparation course', data=df)
```

```python
# Step 14
scatter_general = sns.scatterplot(x=df['reading score'], y=df['writing score'], hue=df['test preparation course'])
```
